## Block Diagram

In this lab we will implement OFDM through enhanced modulation in coding. We will expand the the coherent wireless communications system in lab 08, in which carrier phase information and frequency information are used for data demodulation on the receiver end to acheive OFDM seen in the block diagram below. 




- Data Source

The Data Source module typically receives the data bits from higher-layer applications (e.g., encoded data streams from FFmpeg or Iperf) and delivers the data, in the form of bitarray, to the symbol modulator for digital modulation. When running our system in the BER mode, each bit array is a sequence known to the BER Calculation module in the receiver side so that the BER can be calculated by comparing the transmitted and received data sequences.

- Symbol Modulator

The Symbol Modulator module first combines the payload bits from the Data Source and the preamble bits from the Preamble Generator. Then the complete packet, containing the preamble followed by the payload bits, is modulated into OOK symbols and sent to the OFDM modulation.

#### OFDM Modulator
The symbols recieved from the symbol modulator are processed by the OFDM. The packets are sent to the inverse fast Fourier transform (IFFT) where a cyclic prefix is inserted to copies the end section of IFFT to the beginning of an OFDM symbol. 

- Pulse Shaping

The Pulse Shaping module takes in the symbols and generate digital samples based on the specified pulse shape. The output of this module is an array containing all the samples of the discrete-time baseband signal representing one packet.

- HAL (TX)

The HAL is a module that receives the samples of the BB signal from the pulse shaping module and pushes the samples (in the unit of "buffers") to the ADALM DAC.  

- ADALM DAC
    
The ADALM DAC interpolates the digital samples received from the TX Buffer and generates the corresponding continuous-time waveform at its output using the output sampling rate specified by the user.

- ADALM ADC

The ADALM ADC samples the received analog signal and stores the samples in the receiver buffer. The input sampling rate is specified by the user and the samples are stored in units of "kernel buffers". The size of each kernel buffer and the number of kernel buffers could both be adjusted by the user, but there is an upper limit in terms of the maximum number of samples that could stored in one single kernel buffer.
    
- HAL (RX)

The HAL module requests samples to be transferred from the hardware to the software. Each time a request is made, a number of samples equal to the size of a kernel buffer will be transferred to the HAL module as an array.

- Packet Detection and Frame Syncronization

Upon receiving samples from the Rx Buffer, which contain the BB preamble sequence and data bits, the Frame Sync module performs the frame synchronization by computing the cross-correlation between the received preamble samples and the locally generated preamble samples (also in baseband). From the peak position of the cross-correlation the module identifies the boundary between the preamble and the payload. It has two outputs: the samples of a packet received from the Rx Buffer as well as the array index representing the starting (first) sample of the payload.

- Phase and Frequency Syncronization
The frequency and phase sync modules compute estimates of the frequency and phase offsets of the received baseband signal so that they could be corrected before coherent demodulation & symbol demodulation. The estimation of the carrier frequency offset is obtained via the Fourier transform whereas the carrier phase offset is estimated by extracting the angle/phase from the complex exponential formed by the I and Q channel signals.

- Demodulator (Coherent)

Using the samples with timing, frequency and phase correction applied, the coherent demodulator performs coherent demodulation of the received signal based on the minimum distance / maximum likelihood detection. 


#### OFDM Demodulator
OFDM demodulation includes cyclic prefix removal and FFT operation.

    
- Data Sink

The Data Sink module receives the demodulated bits from the OFDM Demodulator and then sends the bits to higher-layer applications. When running our system in the BER Mode, the higher-layer module is the BER calculator. 
    
- Postprocessing (BER Mode)

As mentioned earlier, if the system is in the BER Mode, a data sequence known to the receiver's BER Calculation Module is transmitted. The known sequence and the demodulated bits are compared on a frame-by-frame as well as bit-by-bit basis to estimate the frame- and bit-error rate.

## System Implementation
    
Our group elected to add a OFDM modulator and OFDM demodulator. OFDM modulation includes IFFT operation and cyclic prefix insertion; OFDM demodulation includes cyclic prefix removal and FFT operation. To determine if the OFDM modulator and demodulator improved the coherent wireless system we used the BER test. The Bit Error Ratio (BER) is used determine the quality of a digital communication system. As an extension of Lab 02, BER is calculated by comparing the transmitted sequence of bits to the received bits and counting the number of errors. The goal of this experient was to get the BER close to 0.
