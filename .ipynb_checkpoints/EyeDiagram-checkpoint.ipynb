```python
## Import key libraries
import math
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
```

```python
## Generate QPSK Symbols with Eb=1.0
N=100
ar=2.0*np.random.randint(2, size=N)-1.0
ai=2.0*np.random.randint(2, size=N)-1.0
a=ar+1j*ai
```

```python
## Overampling factor, i.e., number of samples per symbol period
M=100
```

```python
## Upsampling Filter
p_delta=np.zeros(M) # upsampling filter
p_delta[0]=1.0
```

```python
## DT Rectangular Pulse Shaping Filter
p_rect=np.ones(M)/np.sqrt(M)
```

```python
## Raise Cosine Pulse Shaping Filter
L1=5.  # truncation length in periods

def sinc(x):
    if (x==0):
        s = 1.0
    else:
        s = np.sin(np.pi*x)/(np.pi*x)
    return s

def rcos(t,alpha,T):
    h = np.zeros(np.size(t))
    for i in range(np.size(h)):
        if (alpha==0.0):
            h[i]=sinc(t[i]/T)
        else:
            if (t[i]==T/(2*alpha) or t[i]==-T/(2*alpha)):
                h[i] = np.pi*sinc(1/(2*alpha)) / 4 / T
            else:
                h[i] = sinc(t[i]/T) * np.cos(np.pi * alpha * t[i] / T) / (1-(2*alpha*t[i] / T)**2)
    return h

alpha = 0.25
T = 1.0
Ts = T/M
n=np.arange(2*L1*M+1)
p_rc = rcos(n*Ts-L1*T,alpha,T)
# Normalize for unit energy
p_rc = p_rc / np.sqrt(np.sum(p_rc**2))


plt.stem(p_rc)
```

```python
## Root Raise Cosine Pulse Shaping Filter
L2=5.  # truncation length in periods

def rrcos(t,alpha,T):
    h = np.zeros(np.size(t))
    for i in range(np.size(h)):
        if (alpha==0.0):
            h[i]=sinc(t[i]/T)/np.sqrt(T)
        else:
            if (t[i]==0.0): # See MR Problem A.6
                h[i]= (1/np.sqrt(T)) * (1- alpha + 4*alpha / np.pi)
            else: 
                if (t[i]==T/(4*alpha) or t[i]==-T/(4*alpha)): # See MR Problem A.6
                    h[i] = (alpha/np.sqrt(2*T)) * ((1+2/np.pi)*np.sin(np.pi/4/alpha) + (1-2/np.pi)*np.cos(np.pi/4/alpha))
                else: # MR Equation A.30
                    h[i] = (1/np.sqrt(T)) * (np.sin(np.pi*(1-alpha)*t[i]/T) + (4*alpha*t[i]/T) * np.cos(np.pi*(1+alpha)*t[i]/T)) / (1-(4*alpha*t[i]/T)**2) / (np.pi*t[i]/T)
    return h

alpha = 0.25
T = 1.0
Ts = T/M
n=np.arange(2*L2*M+1)
p_rrc = rrcos(n*Ts-L2*T,alpha,T)
# Normalize for unit energy
p_rrc = p_rrc / np.sqrt(np.sum(p_rrc**2))

plt.stem(p_rrc)
```

```python
## Select a filter, the remainder of the notebook uses only p
p = p_rrc
```

```python
## Upsampling data
u=sig.upfirdn(p_delta,a,M)
n1=np.arange(len(u))
```

```python
plt.stem(n1,np.real(u),markerfmt='D')
plt.xlabel('n')
plt.ylabel('Upsampled')
plt.show()
```

```python
## Transmitter Pulse Shaped Data
s=sig.upfirdn(p,a,M)
n2=np.arange(len(s))
```

```python
plt.plot(n2,np.real(s))
plt.xlabel('n')
plt.ylabel('Pulse Shaped')
plt.show()
```

```python
## Receive Matched Filtered Data
r=sig.upfirdn(p,s)
n3=np.arange(len(r))
```

```python
plt.plot(n3,np.real(r))
plt.xlabel('n')
plt.ylabel('Matched Filter Output')
plt.show()
```

```python
## Transmitter Eye Diagram

delay=int((len(p)-1)/2)
for k in range( math.floor( (len(s)-delay)/(2*M) ) ):
    plt.plot(np.real(s)[delay+k*2*M:delay+(k+1)*2*M],'C0-')
plt.xlabel('n')
plt.ylabel('Transmit Eye Diagram')
plt.show()
```

```python
## Receiver Eye Diagram

delay=len(p)-1

for k in range( math.floor( (len(r)-delay)/(2*M) ) ):
    plt.plot(np.real(r)[delay+k*2*M:delay+(k+1)*2*M],'C0-')
plt.xlabel('n')
plt.ylabel('Receive Eye Diagram')
plt.show()
```

```python
## Receiver Constellation Plot

plt.plot(np.real(r),np.imag(r))
plt.plot(np.real(a),np.imag(a),'x')
plt.xlabel('I')
plt.ylabel('Q')
plt.show
```
